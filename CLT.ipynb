```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
import seaborn as sns
```

## Create population and set parameters

```python
#Create data with exponential distribution
data=np.random.exponential(size=10000)

#Normal distribution
#data=np.random.normal(0, 1, 10000)
```

```python
sns.distplot(data)
```

```python
#Adjust parameters
mu=data.mean()
n_samples=10
samples_of_size=30
mean_of_samples=[]
```

```python
#Find mean of the samples of size
for i in range(n_samples):
    samples=np.random.choice(data,size=samples_of_size)
    mean_of_samples.append(samples.mean())
```

```python
sns.distplot(mean_of_samples)
```
